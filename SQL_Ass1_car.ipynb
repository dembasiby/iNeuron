```python
#q1 . take a data set from attached URL https://archive.ics.uci.edu/ml/datasets/Car+Evaluation
```

```python
import mysql.connector as conn
```

```python
mydb = conn.connect(host='localhost', user='root', passwd='')
```

```python
csor = mydb.cursor()
```

```python
# q2 . create a database called as cardataset
csor.execute('CREATE DATABASE cardataset')
```

```python
# q3 . create a table called as car with a column name given in dataset description
# column names = buying, maint, doors, persons, lug_boot, safety, class
csor.execute('CREATE TABLE cardataset.car(buying VARCHAR(10), maint VARCHAR(10), doors VARCHAR(10), persons VARCHAR(10), lug_boot VARCHAR(10), safety VARCHAR(10), class VARCHAR(10))')
```

```python
# q4 . Dump all the data into car table 

import csv # import csv to work with the csv file
csor.execute('USE cardataset') # switch to the database we will be using from now on.

with open('car.data', 'r') as f:
    data = csv.reader(f, delimiter='\n')
    for row in data:
        values = tuple(row[0].split(','))
        csor.execute('INSERT INTO cardataset.car VALUES(%s, %s, %s, %s, %s, %s, %s)', values)
mydb.commit()
```

```python
# q5 . try to check weather all the data is aviable inside your table or not 
# Two options here:
# 1. We can use csor.execute('SELECT * FROM cardataset.car') and then csor.fetchall()
# 2. Use the Pandas library.

import pandas as pd

pd.read_sql('SELECT * FROM cardataset.car', mydb)
```

```python
# q6 . try to group all the data with COL1 AND count occurences of each and every record based on col1 value 
csor.execute('SELECT buying, COUNT(*) FROM cardataset.car GROUP BY buying')
csor.fetchall()
```

```python
# q7 . Try to filter a record where col 3 value will be 4 .

pd.read_sql('SELECT * FROM cardataset.car WHERE doors=4', mydb)
```

```python
# q8 . Try tox update a col 3 value with 8 whereever you have value equal to 2 

csor.execute('UPDATE cardataset.car SET doors="8" WHERE doors="2"')
mydb.commit()
```
